```python
# 所用的包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Figure_2.3

# 读取
dt = pd.read_csv('D://python/regression/P025b.csv')
#print(dt)

# 提取变量

x1 = dt[['X1']]
y1 = dt[['Y1']]
x2 = dt[['X2']]
y2 = dt[['Y2']]
x3 = dt[['X3']]
y3 = dt[['Y3']]
x4 = dt[['X4']]
y4 = dt[['Y4']]

lrModel1 = LinearRegression()
lrModel2 = LinearRegression()
lrModel3 = LinearRegression()
lrModel4 = LinearRegression()

# 训练模型
lrModel1.fit(x1, y1)
lrModel2.fit(x2, y2)
lrModel3.fit(x3, y3)
lrModel4.fit(x4, y4)

# 拟合值
y1_pred = lrModel1.predict(x1)
y2_pred = lrModel2.predict(x2)
y3_pred = lrModel3.predict(x3)
y4_pred = lrModel4.predict(x4)

# 画图
plt.figure(figsize=(24, 18))
plt.subplot(221)
plt.scatter(x1, y1)
plt.xlabel('X1', fontsize=20)
plt.ylabel('Y1', fontsize=20)
plt.plot(x1, y1_pred, color='red', linewidth=2)
plt.subplot(222)
plt.scatter(x2, y2)
plt.xlabel('X2', fontsize=20)
plt.ylabel('Y2', fontsize=20)
plt.plot(x2, y2_pred, color='red', linewidth=2)
plt.subplot(223)
plt.scatter(x3, y3)
plt.xlabel('X3', fontsize=20)
plt.ylabel('Y3', fontsize=20)
plt.plot(x3, y3_pred, color='red', linewidth=2)
plt.subplot(224)
plt.scatter(x4, y4)
plt.xlabel('X4', fontsize=20)
plt.ylabel('Y4', fontsize=20)
plt.plot(x4, y4_pred, color='red', linewidth=2)
plt.show()
```

```python
# Figure_2.4

# 读取
dt = pd.read_csv('D://python/regression/P027.csv')
#print(dt)

# 提取变量
x = dt[['Units']]
y = dt[['Minutes']]

# 画图
plt.figure(figsize=(24, 18))
plt.scatter(x, y)
plt.xlabel('Units', fontsize=20)
plt.ylabel('Minutes', fontsize=20)
plt.show()
```

```python
# Table_2.6

xi = np.array(dt[['Units']])
yi = np.array(dt[['Minutes']])
index = []
for i in range(1, len(x)+1):
    index.append(i)
i = np.array(index).reshape(len(x), 1)
y1 = yi - np.mean(yi)
x1 = xi - np.mean(xi)
y2 = y1**2
x2 = x1**2
xy = x1 * y1

test_dict = {'i': [int(i[a]) for a in range(0, len(x))],
             'y_i': [int(yi[a]) for a in range(0, len(x))],
             'x_i': [int(xi[a]) for a in range(0, len(x))],
             '(y_i - y_ba)': [np.round(float(y1[a]), 4) for a in range(0, len(x))],
             '(x_i - x_ba)': [int(x1[a]) for a in range(0, len(x))],
             '(y_i - y_ba)^2': [np.round(float(y2[a]), 4) for a in range(0, len(x))],
             '(x_i - x_ba)^2': [int(x2[a]) for a in range(0, len(x))],
             '(y_i - y_ba)(x_i - x_ba)': [np.round(float(xy[a]), 4) for a in range(0, len(x))]
             }
d = pd.DataFrame(test_dict)
d.loc[14] = ['Total', int(np.sum(y)), int(np.sum(x)), np.round(np.sum(y1), 1), int(np.sum(x1)), np.round(np.sum(y2), 4), int(np.sum(x2)), np.round(np.sum(xy), 4)]
print(d)
```

```python
# Figure_2.5

lrModel = LinearRegression()

# 训练模型
lrModel.fit(x, y)

# 拟合值
y_pred = lrModel.predict(x)

# 画图
plt.figure(figsize=(24, 18))
plt.scatter(x, y)
plt.xlabel('Units', fontsize=20)
plt.ylabel('Minutes', fontsize=20)
plt.plot(x, y_pred, color='red', linewidth=2)
plt.show()
```

```python
# Table_2.9

# 数的计算
def sebeta0(y, y_pred, x):
    return ((np.sum((y - y_pred)**2) / (len(x) - int(2)))**0.5) * (int(1) / len(x) + np.mean(x)**2 / np.sum((x - np.mean(x))**2))**0.5

def sebeta1(y, y_pred, x):
    return ((np.sum((y - y_pred)**2) / (len(x) - int(2)))**0.5) / ((np.sum((x - np.mean(x))**2))**0.5)

t0 = lrModel.intercept_/sebeta0(np.array(y), y_pred, np.array(x))

t1 = lrModel.coef_/sebeta1(np.array(y), y_pred, np.array(x))

# 做表
test_dict = {'Variable': ['Constant', 'X'],
             'Coefficient': [float(lrModel.intercept_), float(lrModel.coef_)],
             's.e.': [sebeta0(np.array(y), y_pred, np.array(x)), sebeta1(np.array(y), y_pred, np.array(x))],
             't-test': [float(t0), float(t1)],
             'p-value': ['p0', 'p1']
             }
d = pd.DataFrame(test_dict)
print(d)
```
